# EDA & Visualization Analysis 

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Post_Type', data=df)
plt.title("Distribution of Post Types")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Comments Spread

```python
sns.histplot(df['likes'], kde=True, bins=10, color='skyblue')
plt.title("Likes Distribution")
plt.show()

sns.boxplot(x='comments', data=df)
plt.title("Comments Spread")
plt.show()
```

## Likes by Post Type

```python
sns.boxplot(x='Post_Type', y='likes', data=df)
plt.title("Likes by Post Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Comments vs Likes by Post Type

## Interactive Comments vs Likes

```python
import plotly.express as px

fig = px.scatter(df, x='comments', y='likes', color='Post_Type',
                 hover_data=['Post_id'],
                 title='Interactive Comments vs Likes')
fig.show()
```

## Post Type Distribution

```python
fig = px.pie(df, names='Post_Type', title='Post Type Distribution')
fig.show()
```

## Bubble Chart: Likes vs Comments by Post Type

```python
import plotly.express as px

fig = px.scatter(df, x='comments', y='likes', size='likes', color='Post_Type',
                 hover_data=['Post_id'],
                 title='Bubble Chart: Likes vs Comments by Post Type',
                 size_max=30)
fig.show()
```

## Average Likes & Comments by Post Type

```python
df_grouped = df.groupby('Post_Type')[['likes', 'comments']].mean().reset_index()

fig = px.bar(df_grouped, x='Post_Type', y=['likes', 'comments'],
             barmode='group', title='Average Likes & Comments by Post Type')
fig.show()
```

## Correlation Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df[['likes', 'comments']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

## Treemap of Likes by Post Type

```python
fig = px.treemap(df, path=['Post_Type'], values='likes',
                 title='Treemap of Likes by Post Type')
fig.show()
```

## Likes By ID And Post Type

```python
fig = px.scatter(df, x='Post_id', y='likes', color='Post_Type',
                 trendline='ols', title='Likes By ID And Post Type')
fig.show()
```
